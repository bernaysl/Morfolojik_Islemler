```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/content/j.png',0)
kernel= np.ones((5,5), np.uint8)
```

```python
asindirma =cv.erode(img,kernel, iterations=1) # sadece bir iterasyon yapilacak
genisletme =cv.dilate(img,kernel, iterations=1)
```

```python
basliklar=["img", "asindirma","genisletme"]

resimler=[img,asindirma,genisletme]
```

```python
plt.figure(figsize=(10,12))

for i in range(3):
   #tek satir, uc sutun
  plt.subplot(1,3,i+1), plt.imshow(resimler[i])
  plt.title(basliklar[i])
  plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img_kodluyoruz=cv.imread('/content/kodluyoruz.png')
kernel_10= np.ones((10,10), np.uint8)  #matrisin sayisi ne kadar buyuk olursa etki ettigi piksel sayisi da o kadar buyuk olur
```

```python
asindirma_kodluyoruz=cv.erode(img_kodluyoruz,kernel_10,iterations=1)

genisletme_kodluyoruz=cv.dilate(img_kodluyoruz,kernel_10,iterations=1)
```

```python
images_kodluyoruz=[img_kodluyoruz,asindirma_kodluyoruz,genisletme_kodluyoruz]
```

```python
plt.figure(figsize=(8,11))
for i in range(3):
  plt.subplot(3,1,i+1), plt.imshow(images_kodluyoruz[i])
  plt.title(basliklar[i])
  plt.xticks([]), plt.yticks([])
plt.show()
```
